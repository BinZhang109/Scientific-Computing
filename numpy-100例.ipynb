#### 63. Create an array class that has a name attribute (★★☆)

````python
class NameArray(np.ndarray):
    def __new__(cls,array,name="no name"):
        obj=np.asarray(array).view(cls)
        obj.name=name
        return obj
    def __array_finalize__(self,obj):
        if obj is None:return
        self.info=getattr(obj,'name','no name')
Z=NameArray(np.arange(10),'range_10')
print(Z)
print(Z.name)
````

#### 75. How to compute averages using a sliding window over an array? (★★★)

````python
Z=np.arange(20)
print(Z.mean())
ret=np.cumsum(Z,dtype=float)
print(ret)
print(ret[3:])
print(ret[:-3])
ret[3:]=ret[3:]-ret[:-3]
print(ret[3:])
ret[2:]/3

def moving_average(a,n=3):
    ret=np.cumsum(a,dtype=float)
    ret[n:]=ret[n:]-ret[:-n]
    return ret[n-1:]/n
Z=np.arange(20)
print(moving_average(Z,n=3))
````

#### 76. Consider a one-dimensional array Z, build a two-dimensional array whose first row is (Z\[0\],Z\[1\],Z\[2\]) and each subsequent row is  shifted by 1 (last row should be (Z\[-3\],Z\[-2\],Z\[-1\]) (★★★)

````python
from numpy.lib import stride_tricks
def rolling(a,window):
    shape=(a.size-window+1,window)
    strdes=(a.itemsize,a.itemsize)
    return stride_tricks.as_strided(a,shape=shape,strides=strdes)
A=np.arange(10)
print(rolling(A,9))
````

#### 79. Consider 2 sets of points P0,P1 describing lines (2d) and a set of points P, how to compute distance from each point j (P\[j\]) to each line i (P0\[i\],P1\[i\])? (★★★)

````python
def distance(p0,p1,p):
    T=p1-p0
    L=(T**2).sum(axis=1)
    U=-((p0[:,0]-p[...,0])*T[:,0]+(p0[:,1]-p[...,1])*T[:,1])/L
    U=U.reshape(len(U),1)
    D=p0+U*T-p
    return np.sqrt((D**2).sum(axis=1))
p0=np.random.uniform(-10,10,(10,2))
p1=np.random.uniform(-10,10,(10,2))
p=np.random.uniform(-10,10,(10,2))
print(np.array([distance(p0,p1,p_i) for p_i in p]))
````

````python
class Symtric(np.ndarray):
    def __setitem__(self,key,value):
        i,j=key
        super(Symtric,self).__setitem__((i,j),value)
        super(Symtric,self).__setitem__((j,i),value)
def symetric(Z):
    return np.asarray(Z+Z.T-np.diag(Z.diagonal())).view(Symtric)
Z=np.random.randint(0,10,(5,5))
print(Z)
np.diag(Z.diagonal())
print(Z+Z.T)
s=symetric(Z)
s
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
#1.每个细胞的状态由该细胞及周围八个细胞状态决定
#2.如果一个细胞周围有3个细胞为生，则细胞为生，即该细胞如原先为死 则转为生，若原先为生
#3.如果一个细胞周围有2个细胞为生，则该细胞的生死状态保持不变
#4.在其他情况下，该细胞为死

def iterate(Z):
    #count neighbors
    N=(Z[0:-2,0:-2]+Z[0:-2,1:-1]+Z[0:-2,2:]+
      Z[1:-1,0:-2]+Z[1:-1,2:]+
      Z[2:,0:-2]+Z[2:,1:-1]+Z[2:,2:])
    #apply rules
    birth=(N==3)
    survive=((N==2)|(N==3))
    Z[...]=0
    Z[1:-1,1:-1][birth|survive]=1
    return Z
Z=np.random.randint(0,2,(15,15))
for  i in range(1000):
    Z=iterate(Z)
    print(Z)
````

#### 90. Given an arbitrary number of vectors, build the cartesian product (every combinations of every item) (★★★)

````python
def cartesian(arrays):
    arrays=[np.asarray(a) for a in arrays]
    shape=(len(x) for x in arrays)
    ix =np.indices(shape,dtype=int)
    ix=ix.reshape(len(arrays),-1).T
    for n ,arr in enumerate(arrays):
        ix[:,n]=arrays[n][ix[:,n]]
    return ix
print(cartesian(([1,2,3],[4,5],[6,7])))
# arrays=([1,2,3],[4,5],[6,7])
# arrays=[np.asarray(a) for a in arrays]
# shape=(len(x) for x in arrays)
# ix=np.indices(shape,dtype=int)
# ix=ix.reshape(3,-1).T
# arrays[0][ix[:,0]]
````

#### 98. Considering a path described by two vectors (X,Y), how to sample it using equidistant samples (★★★)?

````python
import matplotlib.pyplot as plt
phi=np.arange(0,10*np.pi,0.1)
print(len(phi))
a=1
x=a*phi*np.cos(phi)
y=a*phi*np.sin(phi)
dr=(np.diff(x)**2+np.diff(y)**2)**.5
r=np.zeros_like(x)
r[1:]=np.cumsum(dr)
print(r.max())
print(r)
r_int=np.linspace(0,r.max(),80)
x_int=np.interp(r_int,r,x)
y_int=np.interp(r_int,r,y)
plt.subplot(133)
plt.plot(x,y,x_int,y_int)
````
